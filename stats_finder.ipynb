```python
import numpy as np
import os
import glob
path_generated = "generated_images/Generated Ground*"
path_gt = "generated_images/Testset_Ground *"
#path_generated = "generated_images_1/Generated Ground*"
#path_gt = "generated_images_1/Testset_Ground *"
path_gt = glob.glob(path_gt)
path_generated = glob.glob(path_generated)
```

```python
from skimage.metrics import structural_similarity as ssim
```

```python
def get_index(path):
    return int(path.split("_")[-1].replace(".png",""))
```

```python
#match path_gt and path_generated by index
# Create dictionaries to map indices to paths
gt_dict = {get_index(path): path for path in path_gt}
generated_dict = {get_index(path): path for path in path_generated}

# Find common indices
common_indices = set(gt_dict.keys()).intersection(set(generated_dict.keys()))
print(len(common_indices))

# Create matched lists
matched_gt = [gt_dict[idx] for idx in sorted(common_indices)]
matched_generated = [generated_dict[idx] for idx in sorted(common_indices)]
```

```python
#read the two images
from skimage.color import rgb2gray
from scipy.stats import pearsonr
from tqdm import tqdm
from PIL import Image
scores = []
for i in tqdm(range(len(matched_gt))):
    gt = Image.open(matched_gt[i])#.convert("RGB")
    generated = Image.open(matched_generated[i])#.convert("RGB")
    
    gt = np.array(gt)
    generated = np.array(generated).max(axis=2)
    
    #to float 
    gt = gt.astype(np.float32)/255
    generated = generated.astype(np.float32)/255
    
    #conver png to grey scale
    #gt = rgb2gray(gt)
    #generated = rgb2gray(generated)
    #score = ssim(gt, generated, data_range=1)
    score = pearsonr(gt.flatten(), generated.flatten())[0]
    scores.append(score)
    
```

```python
#group the scores by every 16 elements
scores_grouped = [min(scores[i:i+16]) for i in range(0, len(scores), 16)]
#find preasonr for top 75
```
